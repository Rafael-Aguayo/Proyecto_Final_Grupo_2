```python
import pandas as pd

# Definir tus columnas y metadatos
data = [
    ["inventoryid", "Identificador único de cada producto o registro en el inventario.", "string"],
    ["store", "Código de la tienda donde se encuentra el producto.", "Int64"],
    ["city", "Ciudad en la que está ubicada la tienda.", "object"],
    ["brand", "Codigo de la marca del producto.", "int64"],
    ["size","Presentacion del producto","string"],
    ["description", "Nombre del producto.", "object"],
    ["size_ml", "Presentación del producto en ML","Int64"],
    ["onhand", "Cantidad actual disponible en inventario.", "int64"],
    ["price", "Precio unitario del producto.", "floa64"],
    ["enddate","Fecha de inventario ","datetime"]
]
# Crear DataFrame
diccionario = pd.DataFrame(data, columns=["columna", "descripcion", "tipo_de_dato"])

# Guardar como CSV
diccionario.to_csv("diccionario_inventario.csv", index=False, sep=";", encoding="utf-8")

print("✅ Diccionario de datos creado: diccionario_inventario.csv")

```
